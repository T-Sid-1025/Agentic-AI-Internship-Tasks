## Problem 1: Mobile Recharge Validation System

```python

def recharge_validation():
    valid_plans = (199, 299, 399, 599)

    while True:
        try:
            amount = float(input("Enter recharge amount: "))
            plan = int(input("Enter plan (199, 299, 399, 599): "))

            if amount < 50:
                print("Recharge amount must be at least ₹50.")
                continue

            if plan not in valid_plans:
                print("Selected plan is not available.")
                continue

            print("Recharge completed successfully!")
            break

        except ValueError:
            print("Invalid input. Please enter numeric values only.")

recharge_validation()
```

## Problem 2: Inventory Reorder Alert System

```python

def inventory_status(product_stock):
    for product_name, quantity in product_stock.items():
        if quantity < 15:
            print(f"{product_name}: Reorder Required")
        else:
            print(f"{product_name}: Stock Sufficient")

inventory_data = {
    "Headphones": 8,
    "Laptop": 25,
    "Mouse": 12,
    "Monitor": 20
}

inventory_status(inventory_data)
```

## Problem 3: Student Result Processing System

```python

def evaluate_student(marks_list):
    total_marks = 0

    for score in marks_list:
        total_marks += score

    average_marks = total_marks / len(marks_list)

    if average_marks >= 50:
        return "Pass"
    else:
        return "Fail"

sample_marks = [55, 40, 60, 70]
print("Result:", evaluate_student(sample_marks))
```

## Problem 4: Cab Fare Estimator with Retry Option

```python

def calculate_cab_fare(distance_km, is_peak_hour):
    base_charge = 50
    rate_per_km = 12

    total_amount = base_charge + (distance_km * rate_per_km)

    if is_peak_hour.lower() == "yes":
        total_amount += total_amount * 0.25

    return total_amount


def cab_fare_system():
    while True:
        try:
            distance = float(input("Enter travel distance (km): "))
            peak = input("Peak hour? (yes/no): ")

            fare = calculate_cab_fare(distance, peak)
            print(f"Total Fare: ₹{fare:.2f}")

            retry = input("Calculate again? (yes/no): ")
            if retry.lower() != "yes":
                break

        except ValueError:
            print("Invalid input. Please enter correct values.")


cab_fare_system()
```

## Problem 5: Employee Attendance Eligibility Checker

```python

def attendance_check(attendance_record):
    present_count = 0

    for status in attendance_record:
        if status == "P":
            present_count += 1

    attendance_percentage = (present_count / len(attendance_record)) * 100

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_sample = ["P", "A", "P", "P", "P", "A", "P", "P"]
print("Eligibility Status:", attendance_check(attendance_sample))
```

## Problem 6: Password Strength Checker

```python

def password_strength(password_text):
    if len(password_text) < 8:
        return "Password too short"

    digit_found = False
    special_found = False
    allowed_special = "@#$"

    for character in password_text:
        if character.isdigit():
            digit_found = True
        if character in allowed_special:
            special_found = True

    if digit_found and special_found:
        return "Strong Password"
    else:
        return "Weak Password"

# Example
print(password_strength("Test@123"))
```
