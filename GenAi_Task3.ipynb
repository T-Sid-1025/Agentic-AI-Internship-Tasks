### Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
a = []
for x in employees.values():
  a.append(x)
b = 0
for i in a:
  if i > b:
    b = i
top_performers = []
for name,score in employees.items():
  if score == b:
    top_performers.append(name)
print(f'Top Performers Eligible for Bonus: {",".join(top_performers)} (Score: {b} )')
```

Problem Statement 2: Search Query Keyword Analysis

```python
e = "Buy mobile phone buy phone online"
e = e.lower()

words = e.split()
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

for word, count in freq.items():
    if count > 1:
        print(f"{word} : {count}")
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}


for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain, count in domain_count.items():
    percent = (count / total) * 100
    print(f"{domain}: {int(percent)}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)

threshold = avg * 1.30

for day in range(len(sales)):
    if sales[day] > threshold:
        print(f"Day {day + 1}: {sales[day]}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

for uid in user_ids:
    if uid in count_ids:
        count_ids[uid] += 1
    else:
        count_ids[uid] = 1


for uid, count in count_ids.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
