## Problem 1: Unique Words in a Sentence


```python
s =  "python is easy and python is powerful"
words = s.split()
unique_words = set(words)
count = len(unique_words)
print("Unique words count:",count)
print("Unique words:",unique_words)
```

# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = max(employees.values())

for name, salary in employees.items():
    if salary == highest_salary:
        print("Highest Salary:", name, "-", salary)
```

## Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

maximum = max(numbers)
minimum = min(numbers)

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# Problem 4: Count Products Above Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_days = attendance.count("P")
total_days = len(attendance)

percentage = (present_days / total_days) * 100

print("Attendance Percentage:", percentage)
```

# Problem 6: Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

# Problem 7: Count Character Frequency

```python
text = "python"

freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

# Problem 9: Check if Key Exists in Dictionary

```python
employees = {"Ravi": 75000, "Anita": 68000}

key = "Ravi"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```
